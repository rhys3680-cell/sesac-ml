## 로지스틱 회귀 구현

### 핵심 아이디어
1. 선형 결합: z = w1·x1 + w2·x2 + ... + b (각 피처 x 가중치의 합)
2. 시그모이드: p = 1 / (1 + e^(-z)) → z를 0~1 확률로 변환
3. 판단: p ≥ 0.5 이면 1, 아니면 0
4. 학습: 예측이 틀린 만큼 가중치를 조금씩 조정 (경사하강법)

### 학습(fit)
트리는 "분기 규칙"을 찾음.
로지스틱 회귀는 **각 피처의 가중치 w를 찾는 것**이 학습이다.
→ 경사하강법으로 "오차를 줄이는 방향으로 w를 조금씩 이동"

### 목표
- 시그모이드·가중치·경사하강법을 numpy로 직접 구현
- 학습이 진행되며 오차(loss)가 줄어드는 것 관찰
- sklearn LogisticRegression과 비교 검증

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    """어떤 실수 z든 0~1 사이 확률로 변환"""
    return 1 / (1 + np.exp(-z))

# z가 -10~10일 때 sigmoid
z = np.linspace(-10, 10, 200)
p = sigmoid(z)

plt.figure(figsize=(7, 4))
plt.plot(z, p)
plt.axhline(0.5, color="red", linestyle="--", label="0.5 (판단 경계)")
```
